```python
%matplotlib inline
import lightkurve as lk
import numpy as np
import pandas as pd
from astropy.io import fits

import os
from pathlib import Path

candidates = pd.read_csv("dataset/candidates.csv")
nuCandidates = pd.DataFrame(columns=candidates.columns)
# nuCandidates.columns = candidates.columns
# print(nuCandidates)

strDir = "C:/Users/jarvi/Desktop/exoplanets/dataset/candidates"

directory = os.fsencode(strDir)

for file in os.listdir(directory):
    for index, row in candidates.iterrows():
        # print(row['kepid'])
        filename = os.fsdecode(file)
        if(str(Path(filename).stem) == str(row['kepid'])):
            print("Match: " + Path(filename).stem)
            nuCandidates.loc[len(nuCandidates)] = row

nuCandidates.head()
nuCandidates.to_csv("dataset/nuCandidates.csv")




```
